```python
import pandas as pd

df = pd.read_csv("./transactions.csv")
```

1. Donnez le montant total des transactions.

```python
print(df['amount'].sum())
```

2. Donnez le montant moyen des transactions.

```python
print(df['amount'].mean())
```

3. Transaction la plus élevée

```python
print(df['amount'].max())
```

4. Type de transaction le plus fréquent

```python
print(df['type'].mode()[0])
```

5. Nombre de transactions par type

```python
print(df['type'].value_counts())
```

6. 10 clients avec le plus de transactions

```python
print(df['nameOrig'].value_counts().head(10))
```

7. Nombre de fraudes

```python
print(df['isFraud'].sum())
```

8. Type de transaction le plus associé à une fraude

```python
print(df[df['isFraud'] == 1]['type'].value_counts().idxmax())
```

9. Somme totale fraudée

```python
print(df[df['isFraud'] == 1]['amount'].sum())
```

 10. Transactions frauduleuses de type CASH_OUT

```python
print(df[(df['isFraud'] == 1) & (df['type'] == 'CASH_OUT')])
```

11. Lignes où oldbalanceDest < newbalanceDest

```python
print(df[df['oldbalanceDest'] < df['newbalanceDest']])
```
