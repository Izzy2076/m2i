```python
from pyspark.sql import SparkSession

spark = SparkSession.builder.master("local").getOrCreate()
sc = spark.sparkContext
```

```python
rddachats = sc.textFile("../Data/achats_clients.csv")
```

```python
rddachats.take(10)
```

```python
first_line = rddachats.first()
rdd = rddachats.filter(lambda ligne: ligne != first_line)
rdd.take(5)
```

```python
rddClients = rdd.map(lambda line : (line.split(",")[0], float(line.split(",")[2])))
rddClients.take(5)
```

```python
montant_clients = rddClients.reduceByKey(lambda a, b : a + b)
result = montant_clients.collect()
```
