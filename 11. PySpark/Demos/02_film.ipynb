```python
from pyspark.sql import SparkSession

spark = SparkSession.builder.master("local").getOrCreate()
sc = spark.sparkContext
```

```python
rddFilm = sc.textFile("../Data/film.data")
```

1. On commence par extraire à partir de chaque ligne la valeur qui nous intéresse :

```python
rddNote = rddFilm.map(lambda line : line.split("\t")[2])
```

```python
rddNote
```

2. On compte le nombre d’éléments par valeur :

```python
result = rddNote.countByValue()
result
```

3. On tri le résultat.

```python
for key, value in sorted(result.items()):
    print(key, value)
```

En une ligne :

```python
result2 = sc.textFile("../Data/film.data") \
.map(lambda line : line.split("\t")[2]) \
.countByValue()

for key, value in sorted(result2.items()):
    print(key, value)
```
