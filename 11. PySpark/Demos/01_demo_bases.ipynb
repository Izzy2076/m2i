```python
# Import de pyspark => pip install pyspark==3.5.6
from pyspark.sql import SparkSession
```

### Création de la session spark :

```python
# Récupération/Création de l'instance spark liée au notebook
spark = SparkSession.builder.master("local").getOrCreate()
# Création d'un context spark pour exécuter des opérations sur le cluster
sc = spark.sparkContext

# def test(x):
#     return x % 2 != 0
```

```python
ma_liste = [3, 6, 8, 10, 25]

rdd = sc.parallelize(ma_liste)
#  Nouvelle RDD avec multiple de 2 de la première => transformation
rdd2 = rdd.map(lambda x: x * 2)
```

```python
result = rdd2.collect()
```
